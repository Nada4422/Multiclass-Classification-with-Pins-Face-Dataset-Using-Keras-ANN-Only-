```python
# Import necessary libraries for data handling, model building, evaluation, and visualization
import numpy as np  # For numerical operations on arrays
import matplotlib.pyplot as plt  # For plotting graphs of training/validation metrics
import seaborn as sns  # For enhancing the visualization of the confusion matrix
import tensorflow as tf  # For deep learning model creation
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # For data augmentation and preprocessing
from tensorflow.keras.models import Sequential  # To define the sequential ANN model
from tensorflow.keras.layers import Dense, Dropout, Flatten, BatchNormalization  # Layers for ANN
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau  # Callbacks to improve training
from sklearn.metrics import classification_report, confusion_matrix  # Evaluation metrics for the model
from sklearn.model_selection import train_test_split  # For splitting data
import cv2  # OpenCV library, though not used in this specific code snippet
import os  # To manage directories
import shutil  # To copy/move files for dataset organization
```

```python
# Define original dataset directory and new split directories for train, validation, and test sets
original_dir = "D:/noody/Deep learning/Lab 3/Multiclass Classification with Pins Face Dataset Using Keras (ANN Only)/archive/105_classes_pins_dataset"
train_dir = "dataset_split/train"
val_dir = "dataset_split/val"
test_dir = "dataset_split/test"

# Make sure the split directories exist by creating them if they don't
for folder in [train_dir, val_dir, test_dir]:
    os.makedirs(folder, exist_ok=True)

# Loop through each class folder in the dataset to split images into train, val, and test sets
for class_folder in os.listdir(original_dir):
    class_path = os.path.join(original_dir, class_folder)  # Path to each class folder
    if not os.path.isdir(class_path):  # Skip non-directory files
        continue
    
    images = os.listdir(class_path)  # List all images in the class folder
    # Split images into 70% train and 30% temp (val + test)
    train_images, temp_images = train_test_split(images, test_size=0.3, random_state=42)
    # Split temp images into 50% val and 50% test, achieving an overall 15% val and 15% test
    val_images, test_images = train_test_split(temp_images, test_size=0.5, random_state=42)
    
    # Move images to their respective directories based on the split
    for img_set, subset_dir in zip([train_images, val_images, test_images], [train_dir, val_dir, test_dir]):
        subset_class_dir = os.path.join(subset_dir, class_folder)  # Path for each subset (train/val/test) per class
        os.makedirs(subset_class_dir, exist_ok=True)
        
        # Copy each image into the correct subset directory
        for img_name in img_set:
            img_path = os.path.join(class_path, img_name)  # Original image path
            shutil.copy(img_path, subset_class_dir)  # Copy image to the destination directory
```

```python
# Data Generators: set up data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values to be between 0 and 1
    rotation_range=20,  # Random rotation of images by up to 20 degrees
    width_shift_range=0.2,  # Horizontal shift of up to 20% of the width
    height_shift_range=0.2,  # Vertical shift of up to 20% of the height
    shear_range=0.2,  # Shearing (slanting) by up to 20%
    zoom_range=0.2,  # Random zooming by up to 20%
    horizontal_flip=True  # Random horizontal flipping
)

# Define data generator for validation and test sets without augmentation
val_test_datagen = ImageDataGenerator(rescale=1./255)  # Normalize pixel values

# Generate training data batches
train_generator = train_datagen.flow_from_directory(
    "dataset_split/train",  # Path to training data
    target_size=(100, 100),  # Resize images to 100x100 pixels
    batch_size=32,  # Number of images per batch
    class_mode='categorical'  # Use categorical labels (multi-class)
)

# Generate validation data batches
validation_generator = val_test_datagen.flow_from_directory(
    "dataset_split/val",  # Path to validation data
    target_size=(100, 100),  # Resize images
    batch_size=32,
    class_mode='categorical'
)

# Generate test data batches
test_generator = val_test_datagen.flow_from_directory(
    "dataset_split/test",  # Path to test data
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical',
    shuffle=False  # Maintain image order for accurate evaluation metrics
)
```

```python
# Obtain the number of classes (unique individuals in the dataset)
num_classes = train_generator.num_classes

# Model Building - Define Artificial Neural Network (ANN) structure
model = Sequential()  # Initialize the sequential model
model.add(Flatten(input_shape=(100, 100, 3)))  # Flatten 100x100 image input with 3 color channels

# First dense layer with ReLU activation, batch normalization, and dropout for regularization
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

# Second dense layer with ReLU activation, batch normalization, and dropout
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

# Third dense layer with ReLU activation, batch normalization, and dropout
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))

# Output layer with softmax activation for multi-class classification, matching the number of classes
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compile the model with Adam optimizer and categorical cross-entropy loss
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Define callbacks to enhance training
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)  # Stop early if no improvement
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-5)  # Reduce learning rate if no improvement

# Train the model, using validation data for monitoring performance during training
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=50,
    callbacks=[early_stopping, reduce_lr]
)
```

```python
# Final evaluation on the test set to assess model performance on unseen data
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")  # Print test accuracy
```

```python
# Generate predictions on test data for detailed evaluation
test_generator.reset()
y_pred = model.predict(test_generator)  # Predict probabilities for each class
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert probabilities to predicted class indices
y_true = test_generator.classes  # True class labels
class_labels = list(test_generator.class_indices.keys())  # Class names

# Print classification report (precision, recall, F1-score per class)
print("Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_labels))
```

```python
# Assuming `conf_matrix` is your confusion matrix, and `class_labels` are your labels
plt.figure(figsize=(50, 50))  # Increase the figure size for better readability
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_labels, yticklabels=class_labels, cbar_kws={'shrink': 0.7})

# Adjust font size and tick label orientation
plt.xticks(rotation=90, fontsize=10)  # Rotate x labels for better fit and adjust fontsize
plt.yticks(rotation=0, fontsize=10)  # Keep y labels horizontal and adjust fontsize

plt.xlabel('Predicted Labels', fontsize=20)
plt.ylabel('True Labels', fontsize=20)
plt.title('Confusion Matrix', fontsize=24)

plt.tight_layout()  # Adjust layout to ensure everything fits
plt.show()
```

```python
# Function to plot training history for loss and accuracy over epochs
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Plot training and validation loss
    ax1.plot(history.history['loss'], label='Train Loss')
    ax1.plot(history.history['val_loss'], label='Validation Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.set_title('Loss Over Epochs')
    ax1.legend()
    
    # Plot training and validation accuracy
    ax2.plot(history.history['accuracy'], label='Train Accuracy')
    ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.set_title('Accuracy Over Epochs')
    ax2.legend()
    
    plt.show()

# Plot training history
plot_training_history(history)
```
